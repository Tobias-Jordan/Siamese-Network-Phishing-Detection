```python
def create_file_links(max_images, save_directory):
    with open(os.path.join(save_directory, 'combined.txt'), 'w') as file:
        for keyword, num_images in max_images.items():
            for folder, other_num_images in max_images.items():
                if folder == keyword:
                    continue
                num_combinations = min(num_images, other_num_images)
                for i in range(1, num_combinations + 1):
                    image_link = f"/content/drive/My Drive/.../{keyword}/{keyword}{i}.png"
                    other_link = f"/content/drive/My Drive/.../{folder}/{folder}{i}.png"
                    line = f"{image_link},{other_link},1\n"
                    file.write(line)

            for i in range(1, num_images + 1):
                image_link1 = f"/content/drive/My Drive/.../{keyword}/{keyword}{i}.png"
                image_link2 = f"/content/drive/My Drive/.../{keyword}/{keyword}{i}.png"
                line = f"{image_link1},{image_link2},0\n"
                file.write(line)

# Beispielaufruf mit den gegebenen max_images
max_images = {
    'alibaba': 10,
    'amazon': 5,
    'amex': 6,
    'apple': 4,
    'chase': 14,
    'facebook': 15,
    'instagram': 6,
    'linkedin': 11,
    'paypal': 6,
    'steam': 6
}

directory = ''
create_file_links(max_images, directory)
```
